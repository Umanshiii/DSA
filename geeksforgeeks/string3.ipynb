```python
# Number of factors

class Solution:
    def countFactors (self, n):
        result=0
        i=1
        while i*i<=n:
            
            if n%i==0:
                if i*i==n:
                    result+=1
                else:
                    result+=2
            
            i+=1
            
        return result
```

```python
#Prime number

class Solution:
    def isPrime(self, n):
        if n<=1:
            return False
            
        fact=set()
        i=1
        while i*i<=n:
            if n%i==0:
                fact.add(i)
                fact.add(n//i)
            i+=1
            
            if len(fact)>2:
                return False
        
        return True
```
