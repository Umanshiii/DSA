```python
#1
#Frequencies in a Limited Array

class Solution:
    def frequencyCount(self, arr):
        #  code here
        n=len(arr)
        result=[0]*n
        
        for i in arr:
            if 1 <= i <= n:
                result[i-1]+=1
            
        return result
```

```python
#2
# Non Repeating Element in an Array

class Solution:
    def firstNonRepeating(self, arr): 
        # code here
        res={}
        
        for i in range(len(arr)):
            res[arr[i]] = res.get(arr[i],0) + 1
        
        for i in range(len(arr)):
            if res[arr[i]]==1:
                return arr[i]
                
        return 0
```
