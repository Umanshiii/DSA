```python
#1
#Max Sum Subarray of size K

class Solution:
    def maxSubarraySum(self, arr, k):
        # code here 
        count=0
        
        for i in range(k):
            count+=arr[i]
        
        maxsum=count
        
        for i in range(k,len(arr)):
            count+=arr[i]
            count-=arr[i-k]
            if count>maxsum:
                maxsum=count
            
        return maxsum
```
