```python
#1
# #sum of array

class Solution:
    def arraySum(self, arr):
        sums=0
        for i in arr:
            sums+=i
            
        return sums
```

```python
#2
# #largest element in an array

class Solution:
    def largest(self, arr):
        # code here
        maximum=arr[0]
        for i in arr:
            if i>maximum:
                maximum=i
        
        return maximum
```

```python
#3
# #mean of an array

class Solution:
    def findMean(self, arr):
        # code here 
        sum=0
        for i in arr:
            sum+=i
        
        return sum//(len(arr))
```

```python
#4
# #Reverse an Array
class Solution:
    def reverseArray(self, arr):
        # code here
        left=0
        right=len(arr)-1
        
        while left<right:
            arr[left],arr[right]=arr[right],arr[left]
            left+=1
            right-=1
            
        return arr
```

```python
#5
# #second largest

class Solution:
    def getSecondLargest(self, arr):
        # Code Here
        if len(arr)<2:
            return -1
        
        largest=-1
        second=-1
        
        for i in arr:
            if i>largest: 
                second=largest
                largest=i
            elif i>second and i<largest:
                second=i
                
        return second
```

```python
#6
#Remove Duplicates Sorted Array

class Solution:
    def removeDuplicates(self, arr):
        # code here
        if len(arr)==0:
            return 0
        
        left=0
        right=1
        while right<len(arr):
            
            if arr[left]!=arr[right]:
                left+=1
                arr[left]=arr[right]
                
            right+=1
            
        return arr[:left + 1]
            
```

```python
#7
#Check if Array is Sorted

class Solution:
    def isSorted(self, arr) -> bool:
        # code here
        for i in range(1,len(arr)):
            if arr[i]<arr[i-1]:
                return False
        
        return True
```

```python
#8
#Max consecutive bits

class Solution:
    def maxConsecBits(self, arr):
        #code here
        cons=1
        maximum=cons
        
        if not arr:
            return 0
        
        for i in range(1,len(arr)):
            if arr[i]==arr[i-1]:
                cons+=1
            else:
                maximum=max(cons,maximum)
                cons=1
                
        maximum=max(cons,maximum)
        
        return maximum
        
```

```python
#9 
#Smallest and Second Smallest

class Solution:
    def minAnd2ndMin(self, arr):
        # code here
        if len(arr)<2:
            return [-1]

        smallest=float('inf')
        second=float('inf')

        for i in arr:
            if i<smallest:
                second=smallest
                smallest=i
            elif i>smallest and i<second:
                second =i

        if second==float('inf'):
            return [-1]

        return [smallest,second]
```
